# Trực quan hóa kết quả PIZZA CLUSTERING

```python
import pandas as pd
# Load the dataset
file_path = 'pizza_clustering_mm.csv'
data = pd.read_csv(file_path)
# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
# Calculate the number of samples in each cluster
cluster_counts = data['CENTROID_ID'].value_counts()
# Display the counts for each cluster
cluster_counts
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Create a scatter plot using seaborn
plt.figure(figsize=(8, 8))
sns.scatterplot(x='fat', y='cal', hue='CENTROID_ID', palette='viridis', data=data, s=50)
# Customize the plot
plt.xlabel('Fat')
plt.ylabel('Calories')
plt.title('Clustering Visualization based on Fat and Calories')
# Add legend title
plt.legend(title='Centroid ID', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
